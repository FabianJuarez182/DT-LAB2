## PRECIOS

```python
pre_super = pd.read_excel('PRECIOS.xlsx', header=22, nrows=12, usecols='A:X')
pre_super
```

```python
pre_regular = pd.read_excel('PRECIOS.xlsx', header=43, nrows=12, usecols='A:X')
pre_regular
```

```python
pre_diesel = pd.read_excel('PRECIOS.xlsx', sheet_name=1, header=1, nrows=12, usecols='A:X')
pre_diesel
```

```python
def convertir(df: pd.DataFrame, name: str) -> pd.DataFrame:
    columns = ['Fecha', name]
    datos = []

    for year in df.columns:
        for i in range(1, 13):
            if year == 'MES/AÑO':
                continue

            year = int(year)
            timestamp = pd.Timestamp(f'{year}-{i}-01')
            temp = df.at[i-1, year]
            datos.append({'Fecha': timestamp, name: temp})

    df_temp = pd.DataFrame(datos, columns=columns)

    return df_temp

df_precio_regular = convertir(pre_regular, 'Regular')
df_precio_super = convertir(pre_super, 'Super')
df_precio_diesel = convertir(pre_diesel, 'Diesel')

df_precios = pd.merge(df_precio_regular, df_precio_super, on='Fecha')
df_precios = pd.merge(df_precios, df_precio_diesel, on='Fecha')

df_precios.set_index('Fecha', inplace=True)
df_precios.sort_index(inplace=True)
df_precios.to_csv('PRECIOS.csv')
df_precios
```
